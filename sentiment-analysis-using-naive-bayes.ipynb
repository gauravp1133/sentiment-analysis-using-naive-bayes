```python
data = pd.read_csv("/kaggle/input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv")
```

```python
df = data.sample(10000)
```

```python
df.head()
```

```python
df['sentiment'] = df['sentiment'].apply(lambda x: 1 if x == 'positive' else 0)
```

```python
df.head()
```

```python
# now we are required to remove any html tags from our reviews

import re
def clear_HTML(review):
    cleanreview = re.compile("<.*?>")
    return re.sub(cleanreview, "", review)
```

```python
df['review'].iloc[5]
```

```python
df['review'] = df['review'].apply(clear_HTML)
```

```python
df['review'].iloc[5]
```

```python
# first, lets convert everything into lowercase

def lower_case(review):
    return review.lower()
df['review'] = df['review'].apply(lower_case)
```

```python
# now lets remove any special characters from the reviews

def remove_special_characters(review):
    cleanreview = re.sub("[^a-zA-Z]", " ", review)
    return cleanreview
df['review'] = df['review'].apply(remove_special_characters)
```

```python
df['review'].iloc[5]
```

```python
# now we need to remove any stopwords

from nltk.corpus import stopwords
stopwords = stopwords.words('english')
print(stopwords)
```

```python
def remove_stopwords(review):
    cleanreview = [word for word in review.split() if word not in stopwords]
    return cleanreview
df['review'] = df['review'].apply(remove_stopwords)
df['review'].iloc[5]
```

```python
df.head()
```

```python
# now we need to perform stemming

from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stemming(review):
    cleanreview = [ps.stem(word) for word in review]
    return cleanreview
df['review'] = df['review'].apply(stemming)
df['review'].iloc[5]
```

```python
# lets join the list back to a string

def join_words(review):
    cleanreview = " ".join(review)
    return cleanreview
df['review'] = df['review'].apply(join_words)
df['review'].iloc[5]
```

```python
# now we will vectorize the data using countvectorizer

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(df['review']).toarray()
y = df['sentiment']
```

```python
X[0]
```

```python
X[0].max()
```

```python
# lets now split the data into training and testing set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# lets apply the naive bayes algorithm to check the sentiment

from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
classifier1 = GaussianNB()
classifier2 = MultinomialNB()
classifier3 = BernoulliNB()
classifier1.fit(X_train, y_train)
classifier2.fit(X_train, y_train)
classifier3.fit(X_train, y_train)
```

```python
# now lets predict the values for each of the model

y_pred1 = classifier1.predict(X_test)
```

```python
y_pred2 = classifier2.predict(X_test)
```

```python
y_pred3 = classifier3.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_pred1))
print(accuracy_score(y_test, y_pred2))
print(accuracy_score(y_test, y_pred3))
```

```python
# now lets create the predictive system, where you put review, and get the sentiment

def predict_sentiment(review):
    review = clear_HTML(review)
    review = lower_case(review)
    review = remove_special_characters(review)
    review = remove_stopwords(review)
    review = stemming(review)
    review = join_words(review)
    review = cv.transform([review]).toarray()
    if classifier3.predict(review) == 1:
        return "Positive"
    else:
        return "Negative"
```

```python
review = input(">>>")

print(predict_sentiment(review))
```
